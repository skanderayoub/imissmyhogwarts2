```python
import json
from duckduckgo_search import DDGS
from urllib.parse import quote
import time

# Load the JSON data
with open('patronus.json', 'r') as file:
    data = json.load(file)

# Extract unique Patronus names
unique_patronus = set()
answers = data['Questions']['Answers']
for answer in answers.values():
    if isinstance(answer, str):
        unique_patronus.add(answer)
    elif isinstance(answer, list):
        for item in answer:
            unique_patronus.add(item)

# Fetch image URLs
patronus_images = {}
with DDGS() as ddgs:
    for patronus in sorted(unique_patronus):
        try:
            # Search for "Patronus [name] Harry Potter" to get relevant images
            query = f"Patronus {patronus} Harry Potter"
            results = list(ddgs.images(query, max_results=1))
            if results:
                patronus_images[patronus] = results[0]['image']
                print(f"✅ {patronus}: {results[0]['image']}")
            else:
                patronus_images[patronus] = None
                print(f"❌ No image found for {patronus}")
            time.sleep(1)  # Avoid rate limiting
        except Exception as e:
            print(f"⚠️ Error fetching {patronus}: {e}")
            patronus_images[patronus] = None

# Save results to a JSON file
with open('patronus_images.json', 'w') as f:
    json.dump(patronus_images, f, indent=2)

print(f"\nSaved {len(patronus_images)} Patronus image URLs to 'patronus_images.json'")
```
